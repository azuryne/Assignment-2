```python
#bytes.decode(encoding="utf-8", errors="strict")

city = "Düsseldorf"  #press u, then press option
utf8_encoded = city.encode('utf-8') 
print(type(utf8_encoded))#bytes
print(utf8_encoded)

decoded_city = utf8_encoded.decode('utf-8') 
print(type(decoded_city)) #str
print(decoded_city)
```

```python
#str.encode(encoding="utf-8",errors="strict")

city = "Düsseldorf"
utf8_encoded = city.encode('utf-8')
print(type(utf8_encoded))
print(utf8_encoded)

decoded_city = utf8_encoded.decode('utf-8')
print(type(decoded_city))
print(decoded_city)
```

```python
#endswith(suffix, beg=0, end=len(string))

#check if ends with suffix

str1 = "this is string example....wow!!!"
suffix = "wow!!!"

print(str1.endswith(suffix))
print(str1.endswith(suffix,20)) #beg with index 20


```

```python
suffix = "is"
print(str1.endswith(suffix, 2, 4)) #begin with index 2 and end with index 4 
print(str1.endswith(suffix, 2, 6))
```

```python
#expandtabs (tabsize = 8 ) - default

str1 = "This is \tstring example....wow!!!"

print("Original string: " + str1)

print("Default expanded tab: " + str1.expandtabs())

print("Double expanded tab: " + str1.expandtabs(16))

print("Random(20) expanded tab: " + str1.expandtabs(20))


```

```python
#find(str, beg=0, end=len(string))

str1 = "this is string example....wow!!!"
str2 = "exam"

print(str1.find(str2))
print(str1.find(str2, 10))
print(str1.find(str2, 40))
```

```python
#index(str, beg=0, end=len(string))

str1 = "this is string example....wow!!!"
str2 = "exam"

print(str1.index(str2))
print(str1.index(str2, 10))
print(str1.index(str2, 40))
```

```python
#isalnum()

#returns True if all characters in str are alphabet/numbers 

str1 = "this2009" #no space in str
print(str1.isalnum())

str2 = "this is string example....wow!!!"
print(str2.isalnum())

str3 = "this is string example wow"
print(str3.isalnum())

str4 = "thisisstringexamplewow"
print(str4.isalnum())
```

```python
#isalpha()
#True -if all chac are alp
#False -if str contains 1 or more non-alp

str1 = "this2009"
print(str1.isalpha())

str2 = "this is string example....wow!!!"
print(str2.isalpha())

str4 = "thisisstringexamplewow"
print(str4.isalpha())
```

```python
#isdigit()
#True -if str only contains digits

str1 = "123456"
print(str1.isdigit())

str2 = "this is string example....wow!!!"
print(str2.isdigit())
```

```python
#islower()
#True - if string has at least 1 cased charac + all lower cased 

str1 = "THIS is string example....wow!!!"
print(str1.islower())

str2 = "this is string example....wow!!!"
print(str2.islower())
```

```python
#isnumeric()
#True - if unicode string contains only numeric chac

str1 = u"this2009"
print(str1.isnumeric())

str2 = u"2345526627178"
print(str2.isnumeric())
```

```python
#isspace()
#True - if str contains ONLY whitespace chac

str1 = "       "
print(str1.isspace())

str2 = "this is string example....wow!!!"
print(str2.isspace())
```

```python
#istitle()
#True - if str is properly "titlecased"

str1 = "This Is String Example....Wow!!!"
print(str1.istitle())

str2 = "this is string example....wow!!!"
print(str2.istitle())
```

```python
#isupper()
#True - if all chac in str are uppercase

str1 = "THIS IS STRING EXAMPLE....WOW!!!"
print(str1.isupper())

str2 = "this is string example....wow!!!"
print(str2.isupper())
```

```python
#join(seq)
# Merge str representation of elements in seq into str, with seperator str

s = "-"
seq = ("a", "b", "c")
print(s.join(seq))

h = "+"
seq1 = ("a", "b", "c")
print(h.join(seq1))
```

```python
#len(string)

str1 = "this is string example....wow!!!"
print("Length of string: ", len(str1))
```

```python
#ljust(width[,fillchar])
#Returns space-padded str with ori str LEFT JUSTIFIED to a total of width columns

str1 = "this is string example....wow!!!"
print(str1.ljust(50, 'o'))

str1 = "this is string example....wow!!!"
print(str1.ljust(40, '+'))
```

```python
#lower()
#Converts all uppercase to lower

str1 = "THIS IS STRING EXAMPLE.....WOW!!!"
print(str1.lower())
```

```python
#lstrip()
#Removes all leading whitespaces in str OR
#Returns a copy of str with all char hace stripped from beg of str

str1 = "        this is string example....wow!!!    "
print(str1.lstrip())

str2 = "@@@@@@@@@this is string example....wow!!!@@@@@@@"
print(str2.lstrip('@'))
```

```python
#maketrans()
#Returns a translation table to be used in translate function 

intab = "aeiou"
outab = "12345"
trantab = str.maketrans(intab, outab) #make table

str1 = "this is string example....wow!!!"
print(str1.translate(trantab))
```

```python
#max(str)
#Returns max alpahbetical chac from str (based on ABC)

str1 = "this is really a string example....wow!!!"
print("Max characters: " + max(str1))

str2 = "this is string example....wow!!!"
print("Max characters: " + max(str2))
```

```python
#min(str)
#Opposite to max

str1 = "this-is-really-a-string-example....wow!!!"
print("Min characters: " + min(str1))

str2 = "this is string example....wow!!!"
print("Min characters: " + min(str2))  #min is whitespace

str3 = "this-is-string-example....wow"
print("Min characters: " + min(str3))  #min is whitespace
```

```python
#replace(old,new[,max])
#Replace all occurences of old in str with new OR 
#at most max occurences in case max is given

str1 = "this is really a string example....wow!!!"
print(str1.replace( "is", "was" ))
print(str1.replace( "is", "was", 3 ))
print(str1.replace( "is", "was", 1 ))
```

```python
#rfind(str, beg=0, end=len(string))
#Same as find(), but search backwards 

str1 = "this is really a string example....wow!!!"
str2 = "is"

print(str1.rfind(str2))
print(str1.rfind(str2, 0, 10))
print(str1.rfind(str2, 10, 0))

print(str1.find(str2))
print(str1.find(str2, 0, 10))
print(str1.find(str2, 10, 0))
```

```python
#rindex(str, beg=0, end=len(string))
#Same as index, but search in backwards string

str1 = "this is really a string example....wow!!!"
str2 = "is"

print(str1.rindex(str2))
print(str1.index(str2))

print(str1.rindex(str2, 0, 10))
print(str1.index(str2, 0, 10))
```

```python
#rjust(width,[,fillchar])
#Returns str RIGHT JUSTIFIED in a string of length width.

str1 = "this is really a string example....wow!!!"
print(str1.rjust(50, 'p'))
print(str1.rjust(50))
```

```python
#rstip()
#Returns a copy of str in which all chars have been stripped from the end of the str 
#(default=whitespace)

str1 = "      this is really a string example....wow!!!     "
print(str1.rstrip())

str2 = "888888888this is really a string example....wow!!!888888"
print(str2.rstrip('8'))
```

```python
#split(str="", num=string.count(str))
#Split str according to delimiter str(space if not provided)
#Returns list of substrings

str1 = "Line1-abcdef \nLine2-abc \nLine4-abcd"

print(str1.split())
print(str1.split(' ', 1))
```

```python
#splitlines(num = string.count('\n'))
#Return splits at all (or num) NEWLINE

str1 = "Line-a b c d e f\nLine2- a b c\nLine4- a b c d"

print(str1.splitlines( ))
print(str1.splitlines( 0 ))
print(str1.splitlines( 3 ))
print(str1.splitlines( 4 ))
print(str1.splitlines( 5 ))
```

```python
#startswith(str, beg=0, end=len(string))

str1 = "this is really a string example....wow!!!"

print(str1.startswith( 'this' ))
print(str1.startswith( 'is', 2 ,4 ))
print(str1.startswith( 'this', 2, 4 ))
```

```python
#strip([chars])

str1 = "0000000this is string example....0000000"

print(str1.strip( '0' ))
```

```python
#swapcase()
#Invert case

str1 = "this is really a string example....wow!!!"

print(str1.swapcase())

str2 = "THIS IS REALLY A STRING EXAMPLE....WOW!!!"

print(str2.swapcase())
```

```python
#title()
#Returns titlecased 

str1 = "this is really a string example....wow!!!"

print(str1.title())
```

```python
#translate(table, deletechars="")

intab = "aeiou"
outtab = '12345'
str1 = "this is really a string example....wow!!!"
trantab = str.maketrans(intab, outtab)

print(str1.translate(trantab))

trantab1 = str.maketrans(intab, outtab, 'xm')
print(str1.translate(trantab1))
```
